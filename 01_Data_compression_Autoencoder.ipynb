```python
import pandas as pd
import numpy as np
from scipy.signal import butter, filtfilt
# Load the text files into a list of numpy arrays
eeg_signals = []
for i in range(1,101):
  if i<10:
    filename = 'F00{}.txt'.format(i) or'N00{}.txt'.format(i)
  elif i<100:
    filename = 'F0{}.txt'.format(i) or 'N0{}.txt'.format(i)
  else:
    filename = 'F{}.txt'.format(i) or 'N{}.txt'.format(i)
  signal = np.loadtxt(filename)
  eeg_signals.append(signal)
for i in range(1,101):
  if i<10:
    filename = 'N00{}.TXT'.format(i)
  elif i<100:
    filename = 'N0{}.TXT'.format(i)
  else:
    filename = 'N{}.TXT'.format(i)
  signal = np.loadtxt(filename)
  eeg_signals.append(signal)
print(eeg_signals)

print(len(eeg_signals))
```

```python
# Convert the list of numpy arrays into a single-column dataframe
df = pd.DataFrame({'signal': eeg_signals})
def butter_bandpass(lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=4):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = filtfilt(b, a, data)
    return y

# Assuming the data is stored in a column 'data' in the dataframe 'df'
fs = 173.61 # Sampling rate
lowcut = 0.5 # Cutoff frequency for low pass filter
highcut = 85 # Cutoff frequency for high pass filter

# Apply the Butterworth filter to each row of the dataframe
df['filtered_signal'] = df['signal'].apply(lambda x: butter_bandpass_filter(x, lowcut, highcut, fs))
print(df)
```

```python
df = df.drop('signal', axis=1)
print(df)
```

```python
def normalize(series):
  mean = series.mean()
  std = series.std()
  return (series - mean) / std

df['normalized_signal'] = df['filtered_signal'].apply(normalize)
print(df)
#df_processed = process_chunk(df, chunk_size=50)
#print(df_processed)
```

```python
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Flatten, Dense, Reshape, UpSampling1D, Cropping1D
from tensorflow.keras.models import Model
from keras.models import Sequential
import numpy as np

# Set the input shape
input_shape = (4097, 1)

# Build the input layer
input_layer = Input(shape=input_shape)

model = Sequential()
model.add(Conv1D(32, 3, activation='relu', padding='same', input_shape=(4097,1)))
model.add(MaxPooling1D( 2, padding='same'))
model.add(Conv1D(8, 3, activation='relu', padding='same'))
model.add(MaxPooling1D(2, padding='same'))
model.add(Conv1D(8,3, activation='relu', padding='same'))
model.add(MaxPooling1D(2, padding='same'))
#model.add(Conv1D(8, 3, activation='relu', padding='same'))
model.add(Flatten())
model.add(Dense(513))
model.add(Reshape((513,1)))
#model.add(Dense(8*512))
#model.add(Reshape((32, 512)))
model.add(Dense(513))
model.add(Conv1D(8, 3, activation='relu', padding='same'))
model.add(UpSampling1D(2))
model.add(Conv1D(8, 3, activation='relu', padding='same'))
model.add(UpSampling1D(2))
model.add(Conv1D(32, 3, activation='relu', padding='same'))
model.add(UpSampling1D(2))
model.add(Conv1D(1,3, activation='relu', padding='same'))
model.add(Cropping1D(cropping=(3,4)))
model.compile(optimizer='rmsprop', loss='mean_squared_error', metrics=['accuracy'])
model.summary()


```
